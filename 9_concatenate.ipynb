```python
import pandas as pd

df = pd.DataFrame({
    'city' : ['mumbai' , 'delhi' , 'bangalore'],
    'temperature' : [32, 45, 30],
    'humidity' : [80 , 60 , 78]
})
df
```

```python
df1 = pd.DataFrame({
    'city' : ['new york' , 'chicago' , 'orland'],
    'temperature' : [21, 14, 35],
    'humidity' : [60 , 65 , 75]
})
df1
```

```python
df3 = pd.concat([df , df1])
df3
```

```python
df3 = pd.concat([df , df1] , ignore_index = True)
df3
```

```python
df3 = pd.concat([df , df1] ,keys = ['India' , 'USA'])
df3
```

```python
df3.loc['India']
```

```python
temp = pd.DataFrame({
    'city' : ['mumbai' , 'delhi' , 'bangalore'],
    'temperature' : [32, 45, 30],
} , index = [0,1,2])
temp

```

```python
humid = pd.DataFrame({
    'city' : ['delhi' , 'mumbai'],
    'humidity' : [23, 41]
} , index = [1,0])
humid
```

```python
df3 = pd.concat([temp , humid])
df3
```

```python
df3 = pd.concat([temp , humid] , axis = 1)
df3
```

```python
s = pd.Series(['Humid' , 'Dry' , 'Rain'] , name = "event")
s
```
